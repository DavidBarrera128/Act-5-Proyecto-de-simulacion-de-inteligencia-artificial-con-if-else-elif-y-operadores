```python
# PROGRAMA: Simulador de IA para Ingeniería en Sistemas
# Reglas estrictas: Solo condicionales, sin funciones, sin ciclos, sin POO.

print("====================================================")
print("🤖 BIENVENIDO AL SISTEMA IA 'INGENIERO EN SISTEMAS' 🤖")
print("====================================================")
print("Hola, soy tu asistente virtual de IA.")
print("Por favor, introduce tu nombre para comenzar:")
nombre = input("> ")

print(f"\nMucho gusto, Ingeniero {nombre}. Vamos a evaluar tu perfil o tu problema actual.")
print("Elige una de las siguientes áreas principales de la carrera:")
print("1. Desarrollo de Software / Programación")
print("2. Redes y Ciberseguridad")
print("3. Inteligencia Artificial y Datos")
opcion_area = input("Digita el número de tu opción (1, 2 o 3): ")

# --- RAMA 1: DESARROLLO DE SOFTWARE ---
if opcion_area == "1":
    print("\n--- SECCIÓN: DESARROLLO DE SOFTWARE ---")
    print("¿Tu sistema actual presenta algún error de ejecución? (si / no)")
    tiene_error = input("> ").lower()

    if tiene_error == "si" or tiene_error == "sí":
        print("¿El error es de sintaxis (código mal escrito) o de lógica (da resultados incorrectos)?")
        tipo_error = input("Escribe 'sintaxis' o 'logica': ").lower()

        if tipo_error == "sintaxis":
            print("\n[IA LOG]: Análisis completado.")
            print("Sugerencia: Revisa los paréntesis, comillas o la indentación del código. El compilador te dice la línea exacta.")
        elif tipo_error == "logica":
            print("\n[IA LOG]: Análisis completado.")
            print("Sugerencia: Utiliza herramientas de depuración (debugging) y revisa el flujo de tus variables paso a paso.")
        else:
            print("\n[IA LOG]: Opción no reconocida. Recuerda revisar la ortografía de tu código.")

    elif tiene_error == "no":
        print("¡Excelente! ¿Quieres optimizar el rendimiento del software? (si / no)")
        optimizar = input("> ").lower()

        if optimizar == "si" and nombre != "":
            print(f"\n[IA LOG]: Configurando entorno para Ingeniero {nombre}...")
            print("Sugerencia: Implementa mejores estructuras de datos y reduce la complejidad algorítmica.")
        else:
            print("\n[IA LOG]: Perfecto. Si el código funciona y no requiere optimización, ¡no lo toques!")
    else:
        print("\n[IA LOG]: Respuesta inválida. El sistema de IA requiere 'si' o 'no'.")

# --- RAMA 2: REDES Y CIBERSEGURIDAD ---
elif opcion_area == "2":
    print("\n--- SECCIÓN: REDES Y CIBERSEGURIDAD ---")
    print("Introduce el porcentaje de tráfico inusual detectado en la red (0 a 100):")
    trafico = int(input("> "))

    if trafico >= 0 and trafico < 40:
        print("\n[IA LOG]: Estado de la red: ESTABLE.")
        print("El tráfico está dentro de los parámetros normales de la empresa.")
    elif trafico >= 40 and trafico <= 75:
        print("\n[IA LOG]: Estado de la red: ADVERTENCIA.")
        print("Sugerencia: Monitorear las IPs de origen. Podría tratarse de una copia de seguridad masiva o un inicio de escaneo.")
    elif trafico > 75 and trafico <= 100:
        print("\n[IA LOG]: 🚨 ALERTA CRÍTICA: POSIBLE ATAQUE DDoS o INTRUSIÓN 🚨")
        print("Sugerencia: Activar el firewall de inmediato, bloquear puertos no esenciales y aislar los servidores principales.")
    else:
        print("\n[IA LOG]: Error. El porcentaje introducido está fuera del rango lógico (0-100).")

# --- RAMA 3: INTELIGENCIA ARTIFICIAL Y DATOS ---
elif opcion_area == "3":
    print("\n--- SECCIÓN: INTELIGENCIA ARTIFICIAL Y DATOS ---")
    print("¿Qué tipo de modelo deseas implementar?")
    print("A. Clasificación de imágenes / Texto")
    print("B. Predicción numérica (Ventas, clima, etc.)")
    modelo = input("Elige la letra de tu opción (A o B): ").upper()

    if modelo == "A":
        print("\n[IA LOG]: Recomendación de arquitectura:")
        print("Deberías utilizar Redes Neuronales Convolucionales (CNN) o modelos basados en Transformers.")
    elif modelo == "B":
        print("\n[IA LOG]: Recomendación de arquitectura:")
        print("Para datos numéricos continuos, inicia con una Regresión Lineal o árboles de decisión (Random Forest).")
    else:
        print("\n[IA LOG]: Error. Opción de modelo no válida.")

# --- MANEJO DE ERRORES PRINCIPAL ---
else:
    print("\n[IA LOG]: Error crítico de entrada.")
    print("La opción seleccionada no corresponde a ninguna rama de la Ingeniería en Sistemas del menú.")

print("\n====================================================")
print("Proceso de simulación finalizado. ¡Éxito en tu proyecto, Ingeniero!")
print("====================================================")
print("programa realizado por David Barrera 1388")
print("====================================================")
```
